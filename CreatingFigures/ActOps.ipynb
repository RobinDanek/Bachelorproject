```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
data = np.loadtxt("D:/Daten mit Änderungen/Physik/Bachelorarbeit/Generated_Data/Original/N20000_a3.0_cosd0.0_e21.0.csv", delimiter=',')
```

```python
plt.figure(figsize=(6,4))
plt.scatter(data[0], np.abs(data[4]), marker='2', s = 2.0)
plt.ylabel("Absolute Opinion")
plt.xlabel("Activity")
```

```python
def lin(x, a, b):
    return a*x + b

popt, pcov = curve_fit(lin, data[0], np.abs(data[4]), p0=[50, 0.0])
acts = np.linspace(0,1,10000)

print(popt)

plt.figure(figsize=(5,3))
plt.scatter(data[0], np.abs(data[4]), marker='2', s = 2.0)
plt.plot(acts, lin(acts, popt[0], popt[1]), color='red')
plt.ylabel("Absolute Opinion")
plt.xlabel("Activity")

plt.savefig("D:\Daten mit Änderungen\Physik\Bachelorarbeit\Figures\ObsAct.png", dpi=500)
```
