```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib.colors import ListedColormap

g = globals()
```

```python
alphas = [2.0, 3.2, 1.2, 1.6, 0.8, 1.8]
cosd = [-0.00, 0.35, 0.50, 0.15, 0.50, 0.15]
classi = [0, 1, 2, 0, 0, 1]

cm = plt.cm.PiYG
cm_data = ListedColormap(['#66c2a5', '#8da0cb', '#fc8d62'])

for i in range (len(alphas)):
    df = pd.read_csv(f"D:\Daten mit Ã„nderungen\Physik\Bachelorarbeit\Generated_Data\Original\\a{alphas[i]}_b5.0_cosd{cosd[i]:.2f}_1.csv", header = None)
    g[f"data{i+1}"] = df.to_numpy()
```

```python
n_rows = 3
n_cols = 2
size = 3

fig, ax = plt.subplots(n_rows, n_cols*2, sharex=False, figsize=(size*n_cols*2, size*n_rows))

for i in range(n_rows):
    for j in range(n_cols):
            ax[i][2*j].hist(g[f"data{n_cols*i+j+1}"][3], bins=250, density=True)
            ax[i][2*j].set_xlim(-10,10)
            ax[i][2*j+1].scatter(g[f"data{n_cols*i+j+1}"][3], g[f"data{n_cols*i+j+1}"][4], alpha=0.1, marker = '2')
            ax[i][2*j+1].set_xlim(-2.5,2.5)
            ax[i][2*j+1].set_ylim(-2.5,2.5)
            ax[i][2*j].grid()
            ax[i][2*j+1].grid()
            ax[i][2*j].set_xlabel("Opinion 1")
            ax[i][2*j+1].set_xlabel("Opinion 1")
            ax[i][2*j+1].set_ylabel("Opinion 2")
            if classi[i*2+j] == 0:
                ax[i][2*j].set_facecolor("#66c2a5")
                ax[i][2*j+1].set_facecolor("#66c2a5")
            elif classi[i*2+j] == 2:
                ax[i][2*j].set_facecolor("#fc8d62")
                ax[i][2*j+1].set_facecolor("#fc8d62")
            elif classi[i*2+j] == 1:
                ax[i][2*j].set_facecolor("#8da0cb")
                ax[i][2*j+1].set_facecolor("#8da0cb")
            ax[i][2*j].set_title(f"$\\alpha$={alphas[2*i+j]}, cos($\delta$)={cosd[2*i+j]}")
            ax[i][2*j+1].set_title(f"$\\alpha$={alphas[2*i+j]}, cos($\delta$)={cosd[2*i+j]}")

plt.tight_layout()
plt.savefig("D:/Daten mit Ã„nderungen/Physik/Bachelorarbeit/Code/Creating_Figures/Agglom_acc.png", dpi=300)
```
